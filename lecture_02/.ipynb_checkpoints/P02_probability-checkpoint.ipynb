## Problem 1: The Gaussian distribution

Let us assume, we measure two quantities $y_1, y_2$ at locations $x_1, x_2$. We know that our observations are independent and each follows a Gaussian distribution with means $\mu_1, \mu_2$ and standard deviations $\sigma_1, \sigma_2$, i.e. $$P(y_1)=\frac{1}{\sqrt{2\pi}\sigma_1}e^{-\frac{(y_1-\mu_1)^2}{2\sigma_1^2}},$$ and $$P(y_2)=\frac{1}{\sqrt{2\pi}\sigma_2}e^{-\frac{(y_2-\mu_2)^2}{2\sigma_2^2}}.$$ We have that $\mu_1=0, \mu_2=2, \sigma_1=0.5, \sigma_2=2$, and the setup is illustrated in the plot below.

(i) What is the joint distribution $P(y_1, y_2)$ of $y_1, y_2$? Write down its analytical form.

(ii) Use `numpy` to draw $n$ samples of $y^i_1, y^i_2$ from this distribution. You can start with $n=100$ and see what happens if you increase $n$.

(iii) Use `matplotlib` to plot histograms of the samples. In particular, plot both the 2D histogram for $y_1, y_2$, as well as the two 1D histograms for $y_1$ and $y_2$.

(iv) Use `matplotlib` to obtain a contour plot from the 2D samples. Overplot it on your 2D historgram and compare.

(v) Derive an analytic expectation for the 1D histograms of $y_1$ and $y_2$ (what distribution are you plotting in this case?). Compare to your findings above. 

(vi) Let us now assume that we are actually not interested in $y_1, y_2$ but in a linear transformation of these variables, $z_1, z_2$ defined as $$\boldsymbol{z}=A\boldsymbol{y},$$ where $$A=\begin{pmatrix}
1 & 0.2  \\
0.2 & 1 
\end{pmatrix}.$$ Using these new samples, repeat problems (i) to (v). What has changed compared to before? 

**Hint:** For (v) you can use the analytic expression for the 2D Gaussian we have discussed in class, i.e. $$\mathcal{N}(\boldsymbol{y}) = \frac{1}{\sqrt{(2\pi)^2|C|}}e^{-\frac{1}{2}(\boldsymbol{y}-\boldsymbol{\mu})^TC^{-1}(\boldsymbol{y}-\boldsymbol{\mu})},$$ with $$C = \begin{pmatrix}
\sigma^2_1 & r\sigma_1\sigma_2  \\
r\sigma_1\sigma_2 & \sigma^2_2

```python
#(ii)
import numpy as np
import matplotlib.pyplot as plt
n = 100
mu1, mu2 = 0, 2
sigma1, sigma2 = 0.5, 2

y1_samples = np.random.normal(loc=mu1, scale=sigma1, size=n)
y2_samples = np.random.normal(loc=mu2, scale=sigma2, size=n)


###(iii) set 100 samples
# 1D Histograms
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(y1_samples, bins=15, density=True, alpha=0.7, color='blue')
plt.title('Histogram of $y_1$f with 100 samples')

plt.subplot(1, 2, 2)
plt.hist(y2_samples, bins=15, density=True, alpha=0.7, color='green')
plt.title('Histogram of $y_2$ with 100 samples')

plt.show()

# 2D Histogram
plt.figure(figsize=(6, 5))
plt.hist2d(y1_samples, y2_samples, bins=15, density=True, cmap='Blues')
plt.colorbar(label='Density')
plt.xlabel('$y_1$')
plt.ylabel('$y_2$')
plt.title('2D Histogram of $y_1$ and $y_2$ with 100 samples')
plt.show()
```

```python
#(iii) set 10000 samples

n = 10000
mu1, mu2 = 0, 2
sigma1, sigma2 = 0.5, 2

y1_samples = np.random.normal(loc=mu1, scale=sigma1, size=n)
y2_samples = np.random.normal(loc=mu2, scale=sigma2, size=n)



# 1D Histograms
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(y1_samples, bins=50, density=True, alpha=0.7, color='blue')
plt.title('Histogram of $y_1$f with 10000 samples')

plt.subplot(1, 2, 2)
plt.hist(y2_samples, bins=50, density=True, alpha=0.7, color='green')
plt.title('Histogram of $y_2$ with 10000 samples')

plt.show()

# 2D Histogram
plt.figure(figsize=(6, 5))
plt.hist2d(y1_samples, y2_samples, bins=50, density=True, cmap='Blues')
plt.colorbar(label='Density')
plt.xlabel('$y_1$')
plt.ylabel('$y_2$')
plt.title('2D Histogram of $y_1$ and $y_2$ with 10000 samples')
plt.show()
```

```python
#(iv)
from scipy.stats import gaussian_kde

# Perform kernel density estimation
xy = np.vstack([y1_samples, y2_samples])
kde = gaussian_kde(xy)
xi, yi = np.mgrid[y1_samples.min():y1_samples.max():100j, y2_samples.min():y2_samples.max():100j]
zi = kde(np.vstack([xi.flatten(), yi.flatten()]))

# Plot 2D Histogram and Contours
plt.figure(figsize=(6, 5))
plt.hist2d(y1_samples, y2_samples, bins=50, density=True, cmap='Blues')
plt.colorbar(label='Density')
plt.contour(xi, yi, zi.reshape(xi.shape), colors='red')
plt.xlabel('$y_1$')
plt.ylabel('$y_2$')
plt.title('2D Histogram with Contour Overlay')
plt.show()
```

## (v)

The analytical expectation for the 1D histograms is simoly the theoretical PDFs of $y_1$ and $y_2$, which are
$$P(y_1)=\frac{1}{\sqrt{2\pi}\sigma_1}e^{-\frac{(y_1-\mu_1)^2}{2\sigma_1^2}},$$ and $$P(y_2)=\frac{1}{\sqrt{2\pi}\sigma_2}e^{-\frac{(y_2-\mu_2)^2}{2\sigma_2^2}}.$$ 

```python
#(v)
from scipy.stats import norm

# Parameters
mu1, mu2 = 0, 2
sigma1, sigma2 = 0.5, 2

# Generate a range of values for plotting the PDFs
y1_values = np.linspace(mu1 - 4*sigma1, mu1 + 4*sigma1, 1000)
y2_values = np.linspace(mu2 - 4*sigma2, mu2 + 4*sigma2, 1000)

# Calculate the PDFs
pdf_y1 = norm.pdf(y1_values, loc=mu1, scale=sigma1)
pdf_y2 = norm.pdf(y2_values, loc=mu2, scale=sigma2)

# Plotting the histogram and PDF for y1
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(y1_samples, bins=50, density=True, alpha=0.7, color='blue', label='Histogram')
plt.plot(y1_values, pdf_y1, 'r-', label='Analytical PDF')
plt.title('Histogram and PDF of $y_1$')
plt.xlabel('$y_1$')
plt.ylabel('Density')
plt.legend()

# Plotting the histogram and PDF for y2
plt.subplot(1, 2, 2)
plt.hist(y2_samples, bins=50, density=True, alpha=0.7, color='green', label='Histogram')
plt.plot(y2_values, pdf_y2, 'r-', label='Analytical PDF')
plt.title('Histogram and PDF of $y_2$')
plt.xlabel('$y_2$')
plt.ylabel('Density')
plt.legend()

plt.tight_layout()
plt.show()

```

### (vi)-(i)
Analytical form of $P(z_{1},z_{2})$

After the linear transformation, $z$ follows a multivariate normal distribution with mean $\mu_{z}=A\mu$ and covariance $C_{z}=ACA^{T}$

mean vector:
$$
\mu=\begin{pmatrix}
\mu_{1} \\
\mu_{2}
\end{pmatrix}=\begin{pmatrix}
0 \\
2 
\end{pmatrix}, \ \ \ \
\mu _{z}=A\mu=\begin{pmatrix}
0.4 \\
2
\end{pmatrix}
$$

Covariance Matrix:

$$
C =\begin{pmatrix}
\sigma_{1}^{2} & 0 \\
0 & \sigma_{2}^{2}
\end{pmatrix}
=\begin{pmatrix}
0.25 & 0 \\
0 & 4
\end{pmatrix}
,$$
$$
C_{z}=ACA^{T}=\begin{pmatrix}
0.41 & 0.85 \\
0.85 & 4.01
\end{pmatrix}.
$$

Joint Distribution:
$$
P(z_{1},z_{2})=\frac{1}{2\pi \sqrt{ \lvert C_{z} \rvert  }}\exp\left( -\frac{1}{2}(z-\mu_{z})^{T}C_{z}^{-1}(z-\mu_{z}) \right)
$$
where $\lvert C_{z} \rvert$ is the determinant of $C_{z}$.

```python
#(vi)-(ii)

# Stack y1 and y2 samples
y_samples = np.vstack([y1_samples, y2_samples])  # Shape: (2, n)

# Transformation matrix A
A = np.array([[1, 0.2], [0.2, 1]])

# Apply the transformation
z_samples = A @ y_samples

# Extract z1 and z2 samples
z1_samples = z_samples[0, :]
z2_samples = z_samples[1, :]


#(vi)-(iii)
# 1D Histograms
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(z1_samples, bins=50, density=True, alpha=0.7, color='purple')
plt.title('Histogram of $z_1$')

plt.subplot(1, 2, 2)
plt.hist(z2_samples, bins=50, density=True, alpha=0.7, color='orange')
plt.title('Histogram of $z_2$')

plt.show()

# 2D Histogram
plt.figure(figsize=(6, 5))
plt.hist2d(z1_samples, z2_samples, bins=50, density=True, cmap='Purples')
plt.colorbar(label='Density')
plt.xlabel('$z_1$')
plt.ylabel('$z_2$')
plt.title('2D Histogram of $z_1$ and $z_2$')
plt.show()

```

```python
#(vi)-(iv)
# Perform kernel density estimation
xy_z = np.vstack([z1_samples, z2_samples])
kde_z = gaussian_kde(xy_z)
xi_z, yi_z = np.mgrid[z1_samples.min():z1_samples.max():100j, z2_samples.min():z2_samples.max():100j]
zi_z = kde_z(np.vstack([xi_z.flatten(), yi_z.flatten()]))

# Plot 2D Histogram and Contours
plt.figure(figsize=(6, 5))
plt.hist2d(z1_samples, z2_samples, bins=50, density=True, cmap='Purples')
plt.colorbar(label='Density')
plt.contour(xi_z, yi_z, zi_z.reshape(xi_z.shape), colors='red')
plt.xlabel('$z_1$')
plt.ylabel('$z_2$')
plt.title('2D Histogram with Contour Overlay for $z$')
plt.show()
```

## Problem 2: The Monty Hall problem (taken from MacKay)

On a game show, a contestant is told the rules as follows:

There are three doors, labeled 1, 2, 3. A single prize has been hidden behind one of them. You get to select one door. Initially your chosen door will not be opened. Instead, the gameshow host will open one of the other two doors, and he will do so in such a way as not to reveal the prize. For example, if you first choose door 1, he will then open one of doors 2 and 3, and it is guaranteed that he will choose which one to open so that the prize will not be revealed.

At this point, you will be given a fresh choice of door: you can either stick with your first choice, or you can switch to the other closed door. All the doors will then be opened and you will receive whatever is behind your final choice of door.
Imagine that the contestant chooses door 1 first; then the gameshow host opens door 3, revealing nothing behind the door, as promised. 

Should the contestant 

(i) stick with door 1, 

(ii) switch to door 2, 

(ii) or does it make no difference?

Explain your solution and reasoning.



We define the relevant events:

$A_1$ : The prize is behind Door 1.

$A_2$ : The prize is behind Door 2.

$A_3$ : The prize is behind Door 3.

$B$ : The host opens Door 3, revealing no prize.

The initial probabilities are uniformly distributed:
$$P(A_i) = \frac{1}{3},i=1,2,3$$

- If the prize is behind Door 1,  $P(B|A_1) = \frac{1}{2}$
- If the prize is behind Door 2,  $P(B|A_2) = 1$
- If the prize is behind Door 3,  $P(B|A_3) = 0$


First, we calculate the denominator $P(B)$:

$$
P(B) = P(B | A_1) \cdot P(A_1) + P(B | A_2) \cdot P(A_2) + P(B | A_3) \cdot P(A_3)=   \frac{1}{2} \cdot \frac{1}{3} + 1 \cdot \frac{1}{3} + 0 \cdot \frac{1}{3}=\frac{1}{2}
$$

Then with the Bayes’ Theorem:
$$
P(A_i | B) = \frac{P(B | A_i) \cdot P(A_i)}{P(B)}
$$
- Probability that the prize is behind Door 1:
$$
 P(A_1 | B) = \frac{P(B | A_1) \cdot P(A_1)}{P(B)} = \frac{\frac{1}{2} \cdot \frac{1}{3}}{\frac{1}{2}} = \frac{1}{3}
$$
- Probability that the prize is behind Door 2:
$$
P(A_2 | B) = \frac{P(B | A_2) \cdot P(A_2)}{P(B)} = \frac{1 \cdot \frac{1}{3}}{\frac{1}{2}} = \frac{2}{3}
$$
Thus, Bayes’ Theorem indicates that **switching doors gives a higher probability of winning**.
